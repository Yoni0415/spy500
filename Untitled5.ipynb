## Análisis Multi-Activo: Patrones y Correlaciones con SPY
En esta sección unificamos los datos de las tecnológicas con el SPY para encontrar señales predictivas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Función para cargar y limpiar cada ticker
def cargar_ticker(archivo, nombre):
    temp = pd.read_csv(archivo, index_col=0, parse_dates=True, header=[0, 1])
    temp.columns = temp.columns.get_level_values(0)
    return temp[['Close']].rename(columns={'Close': nombre})

# Cargar SPY y los nuevos datasets
spy_close = cargar_ticker('spy_historico_completo.csv', 'SPY')
aapl_close = cargar_ticker('AAPL_historico_completo.csv', 'AAPL')
googl_close = cargar_ticker('GOOGL_historico_completo.csv', 'GOOGL')
nvda_close = cargar_ticker('NVDA_historico_completo.csv', 'NVDA')

# Unir todos en un dataframe maestro (solo fechas comunes)
master_df = spy_close.join([aapl_close, googl_close, nvda_close], how='inner')

# Calcular rendimientos logarítmicos para mejor comparación estadística
returns_df = np.log(master_df / master_df.shift(1)).dropna()

print("Matriz de Correlación de Rendimientos Diarios:")
corr_matrix = returns_df.corr()
display(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlación: Tecnológicas vs SPY')
plt.show()
```

### ¿Anticipan las tecnológicas al SPY?
Analizaremos la correlación cruzada con retrasos (lags). Si una tecnológica tiene alta correlación con el SPY de 'mañana', podría servir como indicador adelantado.

```python
lags = range(-5, 6)
lag_results = {}

for col in ['AAPL', 'GOOGL', 'NVDA']:
    corrs = [returns_df['SPY'].corr(returns_df[col].shift(l)) for l in lags]
    lag_results[col] = corrs

plt.figure(figsize=(12, 6))
for ticker, values in lag_results.items():
    plt.plot(lags, values, label=ticker, marker='o')

plt.axvline(0, color='black', linestyle='--', alpha=0.5)
plt.title('Correlación Cruzada (Lead/Lag Analysis)')
plt.xlabel('Días de Retraso (Lag < 0 significa que la acción precede al SPY)')
plt.ylabel('Coeficiente de Correlación')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Nota: Si el pico de la curva está a la izquierda del 0, ese activo tiende a anticipar al SPY.")
```

### Modelo de Anticipación basado en Tecnológicas
Probaremos si el movimiento conjunto de las 3 tecnológicas (AAPL, GOOGL, NVDA) sirve para predecir la dirección del SPY al día siguiente.

```python
# Generar señales basadas en la dirección de las tecnológicas
signals = np.sign(returns_df[['AAPL', 'GOOGL', 'NVDA']])

# Sumamos las señales: Si > 0, la mayoría de tecnológicas subieron. Si < 0, bajaron.
master_df['Tech_Sentiment'] = signals.sum(axis=1)

# La predicción es que el SPY seguirá al Tech_Sentiment al día siguiente
master_df['SPY_Next_Day_Return'] = master_df['SPY'].pct_change().shift(-1)

# Definimos acierto: Si el sentimiento fue positivo y el SPY subió mañana, o viceversa.
def evaluar_acierto(row):
    if row['Tech_Sentiment'] > 0 and row['SPY_Next_Day_Return'] > 0: return 1
    if row['Tech_Sentiment'] < 0 and row['SPY_Next_Day_Return'] < 0: return 1
    if row['Tech_Sentiment'] == 0: return np.nan # Neutral
    return 0

master_df['Hit'] = master_df.apply(evaluar_acierto, axis=1)

accuracy = master_df['Hit'].mean() * 100
print(f"Porcentaje de acierto anticipando el SPY usando AAPL, GOOGL y NVDA: {accuracy:.2f}%")

# Visualización de aciertos por fuerza de la señal
hits_by_strength = master_df.groupby('Tech_Sentiment')['Hit'].mean() * 100
hits_by_strength.plot(kind='bar', color='skyblue', figsize=(10,6))
plt.title('Porcentaje de Acierto según fuerza de la señal Tech')
plt.xlabel('Fuerza de la Señal (-3: Todas bajan, 3: Todas suben)')
plt.ylabel('Precisión (%)')
plt.axhline(50, color='red', linestyle='--')
plt.show()
```

### Análisis de Divergencias como Señales de Reversión
Buscaremos casos donde el sector tecnológico (nuestro proxy de 3 acciones) se desvía significativamente del comportamiento del SPY, lo cual históricamente suele preceder a movimientos correctivos.

```python
# Calculamos la diferencia de rendimiento entre el promedio tech y el SPY
master_df['Tech_Avg_Return'] = returns_df[['AAPL', 'GOOGL', 'NVDA']].mean(axis=1)
master_df['Divergence'] = master_df['Tech_Avg_Return'] - returns_df['SPY']

# Identificamos señales de 'Sobrecompra Tech' (Tech sube mucho más que SPY)
# y 'Sobreventa Tech' (Tech cae mucho más que SPY)
umbral = master_df['Divergence'].std() * 1.5

master_df['Signal_Type'] = 'Neutral'
master_df.loc[master_df['Divergence'] > umbral, 'Signal_Type'] = 'Tech_Bullish_Divergence'
master_df.loc[master_df['Divergence'] < -umbral, 'Signal_Type'] = 'Tech_Bearish_Divergence'

# Analizamos qué pasa con el SPY en los 3 días siguientes a una divergencia fuerte
for i in range(1, 4):
    master_df[f'SPY_Return_T+{i}'] = master_df['SPY'].pct_change().shift(-i)

# Resumen de resultados tras una divergencia
resumen_divergencias = master_df.groupby('Signal_Type')[['SPY_Return_T+1', 'SPY_Return_T+2', 'SPY_Return_T+3']].mean() * 100

print("Rendimiento promedio del SPY (%) tras detectar una divergencia tech:")
display(resumen_divergencias)

# Graficar comportamiento post-divergencia
resumen_divergencias.T.plot(kind='line', marker='o', figsize=(10, 5))
plt.title('Impacto de Divergencias Tech en el SPY (Días posteriores)')
plt.ylabel('Retorno Promedio SPY (%)')
plt.xlabel('Días después de la señal')
plt.axhline(0, color='black', lw=1)
plt.grid(True, alpha=0.3)
plt.show()
```

### Backtesting: Estrategia de Divergencias vs. Inversión Pasiva
Comparamos el rendimiento acumulado de seguir las señales de divergencia frente a simplemente mantener el SPY.

```python
# Definimos la posición: 1 para largo, -1 para corto, 0 para neutral
# Si hay divergencia alcista (tech sube más que SPY), apostamos a que SPY subirá (reversión)
# Si hay divergencia bajista (tech cae más que SPY), apostamos a que SPY bajará
master_df['Strategy_Signal'] = 0
master_df.loc[master_df['Signal_Type'] == 'Tech_Bullish_Divergence', 'Strategy_Signal'] = 1
master_df.loc[master_df['Signal_Type'] == 'Tech_Bearish_Divergence', 'Strategy_Signal'] = -1

# El retorno de la estrategia es la señal de hoy aplicada al retorno de mañana del SPY
master_df['Strategy_Returns'] = master_df['Strategy_Signal'].shift(1) * returns_df['SPY']

# Calculamos retornos acumulados (asumiendo reinversión)
master_df['Cumulative_Passive'] = (1 + returns_df['SPY']).cumprod()
master_df['Cumulative_Strategy'] = (1 + master_df['Strategy_Returns'].fillna(0)).cumprod()

# Visualización
plt.figure(figsize=(14, 7))
plt.plot(master_df['Cumulative_Passive'], label='Pasiva (Buy & Hold SPY)', color='gray', alpha=0.6)
plt.plot(master_df['Cumulative_Strategy'], label='Estrategia Divergencias Tech', color='blue', linewidth=2)
plt.yscale('log')  # Escala logarítmica para ver mejor el crecimiento a largo plazo
plt.title('Rentabilidad Acumulada: Estrategia vs Pasiva ($1 Invertido)')
plt.ylabel('Valor de la Inversión (Log Scale)')
plt.xlabel('Fecha')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()

# Métricas finales
ret_pasiva = (master_df['Cumulative_Passive'].iloc[-1] - 1) * 100
ret_estrategia = (master_df['Cumulative_Strategy'].iloc[-1] - 1) * 100

print(f"Rentabilidad Total Pasiva: {ret_pasiva:.2f}%")
print(f"Rentabilidad Total Estrategia: {ret_estrategia:.2f}%")
```

## Predicción con Machine Learning: Random Forest
En esta sección, utilizaremos las señales previas y datos históricos para entrenar un modelo que aprenda patrones no lineales.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Preparación de Features (X) y Target (y)
# Usaremos retornos pasados de cada activo y la divergencia como predictores
data_ml = master_df[['SPY', 'AAPL', 'GOOGL', 'NVDA', 'Divergence', 'Tech_Sentiment']].copy()

# Creamos lags (retornos de los últimos 3 días)
for col in ['SPY', 'AAPL', 'GOOGL', 'NVDA']:
    for lag in range(1, 4):
        data_ml[f'{col}_Lag_{lag}'] = data_ml[col].pct_change().shift(lag)

# Target: ¿Subirá el SPY mañana? (1 si > 0, 0 en caso contrario)
data_ml['Target'] = (master_df['SPY'].pct_change().shift(-1) > 0).astype(int)

# Limpiar valores nulos generados por los shifts
data_ml = data_ml.dropna()

# Definir X e y
features = [col for col in data_ml.columns if 'Lag' in col] + ['Divergence', 'Tech_Sentiment']
X = data_ml[features]
y = data_ml['Target']

# 2. Split Temporal (No aleatorio para series de tiempo)
split_index = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

# 3. Entrenamiento del Modelo
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, y_train)

# 4. Predicciones
y_pred = model.predict(X_test)

# 5. Evaluación
acc = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo Random Forest en el set de prueba: {acc*100:.2f}%")
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

# Importancia de las variables
plt.figure(figsize=(10, 6))
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh', color='teal')
plt.title('Top 10 Variables más influyentes para el modelo')
plt.show()
```

### Backtesting del Modelo Random Forest
En esta sección aplicamos las predicciones del modelo para simular una estrategia de inversión y compararla con el mercado.

```python
# 1. Preparar los datos de retorno para el set de prueba
test_returns = master_df['SPY'].pct_change().iloc[split_index + (len(master_df)-len(data_ml)):]

# 2. Crear la serie de señales del modelo (y_pred)
# Nota: y_pred ya corresponde cronológicamente a X_test
ml_signals = pd.Series(y_pred, index=X_test.index)

# 3. Calcular retornos de la estrategia
# Si el modelo predice 1, obtenemos el retorno del SPY del día siguiente
strategy_ml_returns = ml_signals * test_returns

# 4. Calcular retornos acumulados
cum_passive_test = (1 + test_returns).cumprod()
cum_strategy_ml = (1 + strategy_ml_returns).cumprod()

# 5. Visualización
plt.figure(figsize=(14, 7))
plt.plot(cum_passive_test, label='Pasiva (Buy & Hold SPY) - Test Set', color='gray', alpha=0.7)
plt.plot(cum_strategy_ml, label='Estrategia Random Forest', color='darkgreen', linewidth=2)
plt.title('Backtest: Random Forest vs Pasiva (Periodo de Prueba)')
plt.ylabel('Valor de la Inversión ($1 inicial)')
plt.xlabel('Fecha')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Métricas Finales
final_ret_passive = (cum_passive_test.iloc[-1] - 1) * 100
final_ret_ml = (cum_strategy_ml.iloc[-1] - 1) * 100

print(f"Rentabilidad Pasiva en el periodo: {final_ret_passive:.2f}%")
print(f"Rentabilidad Estrategia ML en el periodo: {final_ret_ml:.2f}%")
```

### Optimización de Hiperparámetros (Tuning)
Utilizamos búsqueda por rejilla para encontrar la configuración óptima del Random Forest.

```python
from sklearn.model_selection import GridSearchCV

# 1. Definir la rejilla de parámetros a probar
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

# 2. Inicializar el objeto GridSearchCV
# Usamos scoring='accuracy' y 3 folds de validación cruzada temporal
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid,
                           cv=3,
                           n_jobs=-1,
                           verbose=1,
                           scoring='accuracy')

# 3. Ejecutar la búsqueda (usando el set de entrenamiento original)
grid_search.fit(X_train, y_train)

# 4. Obtener el mejor modelo
best_model = grid_search.best_estimator_

print(f"Mejores parámetros encontrados: {grid_search.best_params_}")

# 5. Evaluar el nuevo modelo optimizado
y_pred_opt = best_model.predict(X_test)
acc_opt = accuracy_score(y_test, y_pred_opt)

print(f"\nNueva Precisión (Optimizado): {acc_opt*100:.2f}%")
print(f"Mejora respecto al anterior: {(acc_opt - acc)*100:.2f}%")
```

### Análisis de Rachas: Promedio de Días en Alza y en Baja
Calcularemos cuánto tiempo duran, en promedio, las tendencias antes de cambiar de dirección.

```python
import numpy as np

# Identificar si el día fue alcista (1) o bajista (-1)
df['Direction'] = np.sign(df['Returns']).fillna(0)

# Lógica para agrupar días consecutivos con la misma dirección
df['Streak_ID'] = (df['Direction'] != df['Direction'].shift()).cumsum()

# Calcular la duración de cada racha
streaks = df.groupby('Streak_ID').agg(
    Direction=('Direction', 'first'),
    Duration=('Direction', 'count')
)

# Filtrar rachas de alza (1) y baja (-1)
altas = streaks[streaks['Direction'] == 1]['Duration']
bajas = streaks[streaks['Direction'] == -1]['Duration']

# Resultados
promedio_alta = altas.mean()
promedio_baja = bajas.mean()

print(f"Promedio de días consecutivos al ALZA antes de una caída: {promedio_alta:.2f} días")
print(f"Promedio de días consecutivos a la BAJA antes de una subida: {promedio_baja:.2f} días")

# Visualización de la distribución de las rachas
plt.figure(figsize=(12, 5))
sns.kdeplot(altas, label='Rachas Alza', fill=True)
sns.kdeplot(bajas, label='Rachas Baja', fill=True)
plt.title('Distribución de la Duración de las Rachas (Días)')
plt.xlabel('Días')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### Análisis de Probabilidades (Porcentaje de Acierto)
Calculamos la probabilidad de que, dado que el mercado ha estado en una racha de *N* días, continúe en esa misma dirección al día siguiente.

```python
def calcular_probabilidades_racha(duraciones):
    counts = duraciones.value_counts().sort_index()
    probabilidades = {}
    total_eventos = len(duraciones)

    # Para cada duración N, ¿cuántas veces llegó al menos a N+1?
    for n in range(1, int(duraciones.max())):
        continuaron = len(duraciones[duraciones > n])
        empezaron_n = len(duraciones[duraciones >= n])
        if empezaron_n > 0:
            probabilidades[n] = (continuaron / empezaron_n) * 100

    return pd.Series(probabilidades)

prob_alza = calcular_probabilidades_racha(altas)
prob_baja = calcular_probabilidades_racha(bajas)

print("Probabilidad de que la racha CONTINÚE un día más:")
summary_probs = pd.DataFrame({
    'Días en Racha': prob_alza.index,
    'Acierto Continuidad Alza (%)': prob_alza.values,
    'Acierto Continuidad Baja (%)': prob_baja.reindex(prob_alza.index).values
}).fillna(0)

display(summary_probs.head(10))

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(summary_probs['Días en Racha'], summary_probs['Acierto Continuidad Alza (%)'], marker='o', label='Si hoy subió')
plt.plot(summary_probs['Días en Racha'], summary_probs['Acierto Continuidad Baja (%)'], marker='s', label='Si hoy bajó')
plt.axhline(50, color='red', linestyle='--', alpha=0.5, label='50% (Azar)')
plt.title('Probabilidad de que la tendencia se mantenga un día más')
plt.xlabel('Días actuales en la misma dirección')
plt.ylabel('Probabilidad de continuar (%)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
